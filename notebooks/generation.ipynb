# Steerling-8B: Text Generation (AutoModel)

Load the model via HuggingFace `AutoModel` and generate text using the Steerling generator.

**Requirements:** GPU with >= 18 GB VRAM (A100, A6000, RTX 4090)

```python
import torch
from transformers import AutoModel, AutoTokenizer
from steerling.inference.causal_diffusion import SteerlingGenerator

model_id = "asalam91/steerling-test"

model = AutoModel.from_pretrained(model_id, trust_remote_code=True, torch_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)

generator = SteerlingGenerator.from_model(model, tokenizer, device="cuda")
print(generator)
```

```python
torch.manual_seed(42)

prompt = "The key to understanding artificial intelligence is"
output = generator.generate(prompt, gen_length=128, steps=128, temperature=0.0)
text = generator.decode(output, prompt_len=len(tokenizer.encode(prompt)))

print(f"Prompt: {prompt}")
print(f"Generated: {text}")
```
